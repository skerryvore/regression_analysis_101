```python
import numpy as np
import matplotlib.pyplot as plt

# number of nodes in the mesh
meshes = [11,21,31,61,101]

fig1, ax = plt.subplots(5,1,sharex=True, figsize=(15,6))

# plot cosine function for right most Dirichlet boundary condition as separate plots
for i, mesh in enumerate(meshes):
  x = np.linspace(0,1,mesh)
  ax[i].plot(x,np.cos(6*(3/2)*np.pi*x),label=f"nodes: {mesh:3d} h: {1./mesh:6.4f}")
  ax[i].legend(loc="upper right",fontsize=18)
  ax[i].set_ylabel('cos(f)', fontsize=18)

plt.xlabel('distance', fontsize=18)

plt.savefig('periodic_plots.png', format='png')

fig1.show()





```

```python
fig2, ax = plt.subplots(1,1, figsize=(15,6))

# plot cosine function for right most Dirichlet boundary condition on single plot
for i, mesh in enumerate(meshes):
  x = np.linspace(0,1,mesh)
  ax.plot(x,np.cos(6*(3/2)*np.pi*x),label=f"nodes: {mesh:3d} h: {1./mesh:6.4f}")

plt.xlabel('distance', fontsize=18)
plt.ylabel('Cos(6*(3/2)$\pi$*x)', fontsize=18)
ax.legend(loc='upper right', fontsize=18)

plt.savefig('periodic_plots_all.png', format='png')

fig2.show()
```
